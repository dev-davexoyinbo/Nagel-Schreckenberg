```python
def Nagel_Schreckenberg(road_length, road_density, braking_prob, max_vel, time_iter):
    '''Since I will keep the position of each of the cars with their velocity,
    I will use:
    -1 to denote empty cells and
    0 to denote cells filled with moving cars
    Thus, the number of cars on the road will be the road density multiplied by the road length'''
    start = ([0]*int(road_density * road_length) + [-1]*int((1-road_density) * road_length))
    
    return start

    shuffle(start)
    road_events = [start]

    for time in range(time_iter):
        '''the last vector in the road_events list will be the present state
        while I create a new list to hold the next road event'''
        present_event = road_events[-1]
        next_event = [-1]*road_length
        for car in range(road_length):
            if present_event[car] > -1:
                current_vel = present_event[car]
                '''Acceleration: Increase the current_vel by 1 if it is not at max_vel'''
                next_vel_a = min(current_vel+1, max_vel)
                '''Slowing Down: count the number of empty cells in front of this
                occupied cell and keep the cars' movement within the defined road'''
                count = 1
                while present_event[(car + count) % road_length] < 0:
                    count+=1
                '''Steadily increase the speed of the car to the maximum speed but do not move more than
                the next car'''
                next_vel_sd = min(next_vel_a, count-1)
                '''Randomization: With probability p, reduce speed by 1 (hits break)
                otherwise continue with the current velocity'''
                if uniform(0,1) < braking_prob:
                    next_vel = max(next_vel_sd-1, 0)
                else:
                    next_vel = next_vel_sd
                '''Car Motion: move the car down the cell with the next velocity in the
                next event'''
                next_event[(car+next_vel) % road_length] = next_vel  
        road_events.append(next_event)

    '''Create the road event array over time as an array of 0 and 1 to replace -1 and cells with cars'''
    road_array = np.zeros(shape=(time_iter,road_length))
    for cell in range(road_length):
        for time in range(time_iter):
            if road_events[time][cell] > -1:
                road_array[time,cell] = 1
            else:
                road_array[time,cell] = 0
    '''Average Velocity of the System'''
    vel_sume = 0.0
    vel_count = 0
    for list1 in road_events:
        pos_count = len(list(filter(lambda x: (x >= 0), list1)))
        sum_opera = sum(list(filter(lambda x: (x >= 0), list1)))
        vel_sume+=sum_opera
        vel_count+=pos_count
    if road_density == 0:
        average_vel = 5
    elif road_density == 1:
        average_vel = 0
    else:
        average_vel = vel_sume/vel_count
    return (road_array, average_vel)
```

```python
Nagel_Schreckenberg(100, 0.5, 0.3, 5, 100)
```
